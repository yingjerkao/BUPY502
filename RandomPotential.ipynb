# Single electron in a one-dimensional random potential


In this exercise, we will study the wavefunction of an electron hopping in a one-dimensional random potential on a lattice of length $L$

$$
H=-t\sum_{i}(c^\dagger_i c_{i+1}+\textrm{c.c})-\mu_i c^\dagger_i c_{i}
$$



```julia
using LinearAlgebra
using Plots
```

```julia
const L=200
```

```julia
function H_random(V0::Float64, L::Int64)
    H=zeros(L, L)
    for i in 1:L
        H[i,(i+L)%L+1]=-1.0
        H[(i+L)%L+1,i]=-1.0
        H[i,i]=-V0*rand()
    end
    return H
end
```

```julia
function H_gaussian(V0::Float64, L::Int64)
    H=zeros(L, L)
    for i in 1:L
        H[i,(i+L)%L+1]=-1.0
        H[(i+L)%L+1,i]=-1.0
        H[i,i]=1-exp(-(i-L/2)^2/L^2)-V0*rand()
    end
    return H
end
```

```julia
H1=H_random(0.01,L)
```

```julia
E1=eigvals(H1)
```

```julia
ψ1=eigvecs(H1)
```

```julia
plot([i for i in 1:L],ψ1[:,1].^2 , label="\$|\\Psi|^2\$")
plot!([i for i in 1:L], diag(H1) , label="\$\\mu\$")
```

```julia
H2=H_gaussian(0.01,L)
```

```julia
ψ2=eigvecs(H2)
```

```julia
plot([i for i in 1:L],ψ2[:,1].^2, label="\$|\\Psi|^2\$")
plot!([i for i in 1:L], diag(H2) , label="\$\\mu\$")
```
